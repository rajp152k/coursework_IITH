```python
import pandas as pd
import numpy as np
import random
from itertools import product
from typing import Callable
np.random.seed(1729)
```

## Board

```python
test_index = lambda i: i<=2 and i>=0

class TicTacToeBoard:
    def __init__(self):
        self.row = [' ']*3
        self.board = [self.row, self.row.copy(),self.row.copy()]
        self.left = set(range(9))
        self.agent_char = 'X'
        self.opponent_char = 'O'
        self.tie_char = 'T'
        self.agent = set()
        self.opponent = set()
        self.action_space = range(9)
        # defining groups of sub-states the signify a win for a player
        self.terminators =   ({0,1,2}, {3,4,5}, {6,7,8},
                              {0,3,6}, {1,4,7}, {2,5,8},
                              {0,4,8}, {2,4,6})
        
        self.print = self.__repr__
        
    def reset(self):
        # to be executed before each new game
        self.__init__()
        
    def act(self, i: int, j: int):
        '''
        Mover for agent, i is the row and j is the col in which to place the agent character
        should only be called by the agent for correctness of state info (player sets)
        '''
        assert self.board[i][j] == ' ' and test_index(i) and test_index(j), f"invalid move requested by agent: {i,j,board}"
        self.board[i][j] = self.agent_char
        self.left = self.left - {i*3 + j}
        self.agent.add(i*3 + j)
        
    def oppose(self, i: int, j: int):
        '''
        Mover for opponent, i is the row and j is the col in which to place the opponent character
        should only be called by the opponent for correctness of state info (player sets)
        '''
        assert self.board[i][j] == ' ' and test_index(i) and test_index(j), "invalid move requested by opponent"
        self.board[i][j] = self.opponent_char
        self.left = self.left - {i*3 + j}
        self.opponent.add(i*3 + j)
        
    def __getitem__(self, i: int):
        return self.board[i]
    
    def __repr__(self):
        '''
        for printing the board: callable by print(board: TicTacToeBoard)
        '''
        b = lambda i,j: self.board[i][j]
        repr = '\n'
        for i in range(3):
            for j in range(3):
                if j == 2:
                    repr += f' {b(i,j)} \n'
                    continue
                repr += f' {b(i,j)} |'
            if i == 2:
                continue
            repr += '-'*11 + '\n'
        return repr
    
    def check_term(self, contender: int):
        '''
        terminal check specific to whether the contender won 
        0 for agent
        1 for opponent
        '''
        contender = self.agent if contender == 0 else self.opponent
        for trmntr in self.terminators:
            if all(place in contender for place in trmntr):
                return True
        return False
    
    def is_terminal(self):
        '''
        Termination semantics based upon return:
        1. False : is non-terminal
        2. 'X' : agent won
        3. 'O' : opponent won
        4. 'T' : Tie
        '''
        if self.check_term(0):
            return 'X'
        elif self.check_term(1):
            return 'O'
        elif not self.left:
            return 'T'
        else:
            return False 
        
    def state_hash(self):
        '''
        Generates a unique hashable string for the current state of the board
        used to quickly index into the Q-table of the agent for updates
        '''
        hash_str = ''
        for i in range(3):
            for j in range(3):
                hash_str += self.board[i][j]
        return hash_str
```

## Opponent 

```python
class Opponent():
    def __init__(self,strat:int):
        '''
        returns an Opponent that plays according to the given strategy
        0 for random
        1 for safe
        '''
        self.move = self.random_move if strat == 0 else self.safe_move
        self.execute = lambda action,board: board.oppose(action//3,action%3)
        
    def random_move(self,board: TicTacToeBoard):
        '''
        choose a move at random for the opponent('O') given the current state
        returns False if board is full
        '''
        if not board.left:
            return False
        action = random.sample(board.left,1)[0]
        self.execute(action,board)
        return True
        
    def safe_move(self,board: TicTacToeBoard):
        '''
        choose a safe move (heuristics-based) for opponent('O') given the current state
        else defer to a random move if unavailable
        '''
        win = self.check_win(board)
        if win:
            self.execute(win,board)
            return True
        block = self.check_block(board)
        if block:
            self.execute(block,board)
            return True
        self.random_move(board)
        
        
    def check_win(self,board: TicTacToeBoard):
        '''
        check if 'O' can win and return corresponding action if possible
        False otherwise
        '''
        for termination in board.terminators:
            if len(board.opponent.intersection(termination)) == 2:
                potential_move = (termination - board.opponent).pop()
                if potential_move in board.left:
                    return potential_move
        return False
        
    def check_block(self,board: TicTacToeBoard):
        '''
        check if 'X' can be blocked and return corresponding action if possible
        False otherwise
        '''
        for termination in board.terminators:
            if len(board.agent.intersection(termination)) ==2:
                potential_block = (termination - board.agent).pop()
                if potential_block in board.left:
                    return potential_block
        return False
```

## Agent

### Q table 

```python
states = [''.join(t) for t in list(product(['X','O',' '],repeat=9))]
action_space = range(9)

tabQ = pd.DataFrame(index = states, columns = action_space,data = np.random.uniform(-0.1,0.1,(len(states),9)))
countQ = pd.DataFrame(index = states, columns = action_space,data = np.zeros((len(states),9)))
tabQ
```

```python
class Agent():
    def __init__(self,basetabQ: pd.DataFrame,basecountQ: pd.DataFrame,gamma: float):
        '''
        returns an agent with a freshly initialised table for Q-values
        '''
        self.tabQ = basetabQ.copy()
        self.countQ = basecountQ.copy()
        self.gamma = gamma
        
        self.reward_tab = {  'X':1,    # agent won
                             'O':-1,   # agent lost
                             'T':0,   # tie
                              False: 0} # non-terminal state}
        
        self.execute = lambda action,board: board.act(action//3,action%3)
        
    def fetch_reward(self, res):
        '''
        returns reward when queried by the terminal nature of a state
        '''
        return self.reward_tab[res]
        
    def epsilon_greedy(self,board:TicTacToeBoard, epsilon: float):
        '''
        returns the epsilon greedy action recommended for the current state dependent
        '''
        if np.random.uniform(0.,1.) > epsilon:
            return self.tabQ.loc[board.state_hash()][board.left].idxmax()
        else:
            return np.random.choice(list(board.left))
        
    def greedy(self,board: TicTacToeBoard):
        '''
        returns the fully greedy action based on the current Q-value table
        '''
        return self.tabQ.loc[board.state_hash()][board.left].idxmax()
        
    def update_tabQ(self, s, action, s_prime, reward, alpha: float):
        '''
        update the Q-table for a s-a-r-s 4-tuple using the learning rate as alpha
        '''
        correction = reward + self.gamma * max(self.tabQ.loc[s_prime]) - self.tabQ.loc[s][action]
        self.tabQ.loc[s][action]+= alpha * correction
        
    def update_countQ(self,s,action):
        '''
        increment number of visits for current state-action pair
        '''
        self.countQ.loc[s][action] += 1
        
    def update_terminal_tabQ(self,terminal_s):
        '''
        setting Q-value for terminal states whenever encountered
        '''
        self.tabQ.loc[terminal_s] = 0.
```

## One Game

```python
def execute_game(agent: Agent, opponent: Opponent, board: TicTacToeBoard, epsilon: float, alpha: float,test: bool):
    '''
    Execute one game of Tic Tac Toe given an agent and an opponent
    '''
    opponent_first = np.random.uniform(0.,1.) <= 0.5 # coin toss for first player
    board.reset() # reset board
    done = False # termination controller

    if opponent_first:
        opponent.move(board)
    while not done:
        s = board.state_hash()
        if not test:
            # training 
            action = agent.epsilon_greedy(board, epsilon)
        else:
            # testing
            action = agent.greedy(board)
        agent.execute(action,board)
        s_mid = board.state_hash()
        end = board.is_terminal()
        # if terminal state, will update Q-table
        if end == board.agent_char or end == 'T':
            # checking win for agent or tie (loss not possible)
            reward = agent.fetch_reward(end)
            if not test:
                # training updates
                agent.update_tabQ(s, action, s_mid, reward, alpha)
                agent.update_countQ(s,action)
                agent.update_terminal_tabQ(s_mid)
            done = True
            result = end
            continue
        opponent.move(board)
        s_prime = board.state_hash()
        end = board.is_terminal()
        reward = agent.fetch_reward(end)
        if not test:
            # s_prime will be the final state for agent instead of s_mid if no termination in the middle
            agent.update_tabQ(s, action, s_prime, reward,alpha)
            agent.update_countQ(s,action)
        if end == board.opponent_char or end == 'T':
            if not test:
                agent.update_terminal_tabQ(s_prime)
            done = True
            result = end
            
    return result
```

## Tester

```python
def test_agent(agent: Agent, summon_opp: Callable, board: TicTacToeBoard, num_games: int):
    '''
    executes several games with completely greedy agent against randomly selected opponents for num_games and returns results
    '''
    tally = [0,0,0] # wins,ties,losses
    result_hash = {board.agent_char: 0,
                   board.opponent_char: 2,
                   board.tie_char: 1}
    
    def update_tally(result: str):
        tally[result_hash[result]] += 1
    
    # note that testing is done against a randomly selected opponent
    for game in range(num_games):
        update_tally(execute_game(agent,summon_opp(fixed = False),board,0,0,True))
    
    return tally
```

# Init 

```python
from tqdm import tqdm

num_games = 10000
test_every = 200
num_test_games = 100
num_final_test_games = 1000

alpha_update_every = 500
alpha_decay_factor = 0.9

eps_update_every = 500
eps_decay_factor = 0.9

gamma = 0.8

safe_opponent = Opponent(1) 
random_opponent = Opponent(0)
# storing opponent objects to avoid creating new ones
# note that they are stateless and hence can be reused unlike an agent

board = TicTacToeBoard()

def summon_opponent(fixed: False,strat = 0):
    '''
    returns an opponent:
    if not fixed, returns a randomly selected opponent,
    otherwise, returns one according to the requested strategy
        0 for random 
        1 for safe
    '''
    if not fixed:
        return random.choice([safe_opponent,random_opponent])
    else:
        if strat == 0:
            return random_opponent
        else:
            return safe_opponent
```

```python
import matplotlib.pyplot as plt

def present_results(test_wins,
                    test_ties,
                    test_losses,
                    test_idxs,
                    final_wins,
                    final_ties,
                    final_losses,
                    descr:str):
    fig,ax = plt.subplots(nrows= 2,ncols=1,figsize=(15,15))
    plot_train_results(ax[0],test_wins,test_ties,test_losses,test_idxs,descr)
    plot_final_results(ax[1],final_wins,final_ties,final_losses,descr)
    fig.show()
    
def plot_train_results(ax,test_wins,test_ties,test_losses,test_idxs,descr):
    ax.set_title(f'{descr}: train results')
    ax.plot(test_idxs,test_wins,'g-',label="wins")
    ax.plot(test_idxs,test_ties,'y-',label="ties")
    ax.plot(test_idxs,test_losses,'r-',label="losses")
    ax.set_ylim(0,num_test_games)
    ax.legend()
    
def plot_final_results(ax,final_wins,final_ties,final_losses,descr):
    ax.set_title(f'{descr}: final results')
    ax.pie([final_wins,final_ties,final_losses],labels=["wins","ties","losses"],autopct='%.1f%%')
```

### Random Opponent

```python
num_test_wins = []
test_idxs = []
test_wins = []
test_ties = []
test_losses = []
fixed_opponent = True
strat = 0
opponent = summon_opponent(fixed_opponent, strat)

alpha = 0.9
epsilon = 0.5

agent_ro = Agent(tabQ,countQ, gamma)

for game_no in tqdm(range(1,num_games+1)):
    execute_game(agent_ro,opponent,board,epsilon,alpha,False)
    
    if not(game_no % alpha_update_every):
        alpha *= alpha_decay_factor
    if not(game_no % eps_update_every):
        epsilon *= eps_decay_factor
    
    if not (game_no % test_every):
        wins,ties,losses = test_agent(agent_ro,summon_opponent,board,num_test_games)
        test_wins.append(wins)
        test_ties.append(ties)
        test_losses.append(losses)
        test_idxs.append(game_no)
        
wins,ties,losses = test_agent(agent_ro,summon_opponent,board,num_final_test_games)
present_results(test_wins,test_ties,test_losses,test_idxs,wins,ties,losses,
               'Random Opponent')
```
