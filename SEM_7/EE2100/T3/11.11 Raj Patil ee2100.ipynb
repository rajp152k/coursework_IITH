```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
get_row = lambda t,y : np.array([[1,t,t**2,-y*t,-y*t**2]])
```

```python
ts = [1,2,3,4,5]
ys = [2,5,9,-1,-4]
```

```python
A = np.array([get_row(t,y) for t,y in zip(ts,ys)])
b = np.array(ys)
```

Solving for $Ax = b$

```python
coeffs = np.linalg.solve(A,b)
```

```python
def gen_rational_f(coeffs):
    c1,c2,c3,d1,d2 = coeffs
    return lambda t : (c1 + c2*t + c3*t**2)/(1 + d1*t + d2*t**2)
```

```python
f = gen_rational_f(coeffs)
```

```python
xs = np.arange(0,6,0.01)
plt.plot(xs, [f(x) for x in xs])
plt.scatter(ts,[f(t) for t in ts])
for t in ts:
    plt.axvline(x=t,c="g",ls="--")
for y in ys:
    plt.axhline(y=y,c="g",ls="--")
plt.title("Estimating Rational f")
plt.show()
```
