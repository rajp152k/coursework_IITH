```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline, BSpline
```

## Num simulations 

```python
df_call = pd.read_csv("num_sim_call.csv").drop(0)
df_put = pd.read_csv("num_sim_put.csv").drop(0)
```

```python
x = range(100,10000,100)
```

```python
plt.figure(figsize= (15,10))
plt.plot(x,df_call['ste'],label="call_ste")
plt.plot(x,df_put['ste'],label="put_ste")
plt.legend()
plt.xlabel("Number of simulations")
plt.ylabel("Standard Error")
plt.title("Error with Number of Simulations")
plt.show()
```

```python
df_call
```

```python
plt.figure(figsize= (15,10))

plt.plot(x,df_call['call_price'],label="call_price")
plt.plot(x,df_call['black_scholes_call'],"g--",label="black_scholes_call")

plt.legend()
plt.xlabel("Number of simulations")
plt.ylabel("Option Price")
plt.title("Call Price Evolution with Number of Simulations")
plt.show()
```

```python
plt.figure(figsize= (15,10))

plt.plot(x,df_put['put_price'],label="put_price")
plt.plot(x,df_put['black_scholes_put'],"g--",label="black_scholes_put")


plt.legend()
plt.xlabel("Number of simulations")
plt.ylabel("Option Price")
plt.title("Put Price Evolution with Number of Simulations")
plt.show()
```

## Antithetic 

```python
df_call = pd.read_csv("antithetic_call.csv").drop(0)
df_put = pd.read_csv("antithetic_put.csv").drop(0)
```

```python
df_call
```

```python
plt.figure(figsize= (20,10))

plt.plot(x,df_call['std'],label="call_std")
plt.plot(x,df_call['std_antithetic'],label="call_antithetic_std")

plt.legend()
plt.xlabel("Number of simulations")
plt.ylabel("Standard Deviation")
plt.title("Call - Effect of Antithetic Variate")
plt.show()
```
